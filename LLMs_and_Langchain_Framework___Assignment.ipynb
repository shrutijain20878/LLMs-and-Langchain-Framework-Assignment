**Question 1: What are Large Language Models (LLMs) and how do they function?**

Large Language Models (LLMs) are advanced artificial intelligence models designed to understand and generate human-like text. They are typically built using deep learning architectures such as the Transformer Architecture and are trained on massive amounts of text data from books, websites, and articles. LLMs learn patterns, grammar, context, and relationships between words during training. When given a prompt, the model predicts the most likely next words based on what it has learned, allowing it to generate meaningful responses, answer questions, translate languages, summarize text, and perform many other natural language processing tasks.

**Question 5: Compare and contrast Langchain and LamaIndex. What unique problems** **does each solve?**

LangChain and LlamaIndex are both frameworks used to build applications with Large Language Models, but they focus on different problems. LangChain helps developers create complex LLM workflows by connecting models with tools, APIs, memory, and agents to build applications like chatbots and automation systems. In contrast, LlamaIndex focuses mainly on connecting LLMs with external data sources such as documents, PDFs, and databases, enabling efficient data retrieval and question answering. In simple terms, LangChain manages the application logic and workflows, while LlamaIndex specializes in data indexing and retrieval for LLMs.

```python
# Question 6: Implement a basic Langchain pipeline using OpenAI’s LLM to answer questions based on a user input prompt.
!pip install langchain openai

from langchain.llms import OpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
import os

os.environ["OPENAI_API_KEY"] = ""

llm = OpenAI(temperature=0.7)

prompt = PromptTemplate(
    input_variables=["question"],
    template="Answer the following question clearly:\nQuestion: {question}\nAnswer:"
)

chain = LLMChain(llm=llm, prompt=prompt)

user_question = input("Enter your question: ")

response = chain.run(user_question)

print("Answer:", response)
```

```python
# Question 7: Integrate Langchain with a third-party API (e.g., weather, news) and show how responses can be generated via LLMs.
!pip install langchain openai requests

import os
import requests
from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

os.environ["OPENAI_API_KEY"] = " "

WEATHER_API_KEY = "your_weather_api_key"

def get_weather(city):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={WEATHER_API_KEY}&units=metric"
    response = requests.get(url).json()

    weather = response["weather"][0]["description"]
    temp = response["main"]["temp"]

    return f"The current weather in {city} is {weather} with a temperature of {temp}°C."


llm = ChatOpenAI(temperature=0.5)


prompt = PromptTemplate(
    input_variables=["weather_data"],
    template="Explain the following weather information in a friendly way:\n{weather_data}"
)

chain = LLMChain(llm=llm, prompt=prompt)

city = input("Enter city name: ")

weather_data = get_weather(city)

response = chain.run(weather_data)

print("LLM Response:")
print(response)
```

```python
# Question 9: Demonstrate combining Langchain with LamaIndex to create a simple document-based Q&A chatbot.
!pip install langchain llama-index openai

import os
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader
from langchain.chat_models import ChatOpenAI
from langchain.chains import ConversationalRetrievalChain

os.environ["OPENAI_API_KEY"] = ""

documents = SimpleDirectoryReader("data").load_data()

index = VectorStoreIndex.from_documents(documents)

retriever = index.as_retriever()

llm = ChatOpenAI(temperature=0)

qa_chain = ConversationalRetrievalChain.from_llm(llm, retriever)

chat_history = []

while True:
    query = input("Ask a question: ")

    result = qa_chain({"question": query, "chat_history": chat_history})

    print("Answer:", result["answer"])

    chat_history.append((query, result["answer"]))
```
